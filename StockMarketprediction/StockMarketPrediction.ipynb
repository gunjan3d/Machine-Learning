```python
df = pd.read_csv('2021to2023.csv')
df['Open ']
df['Open'] = df['Open ']
df = df.drop('Open ',axis = 1)
df['High'] = df['High ']
df = df.drop('High ',axis=1)
df['Low'] = df['Low ']
df = df.drop('Low ',axis=1)
df['Close'] = df['Close ']
df = df.drop('Close ',axis=1)
df
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

X = df[['Open']]  # Reshape to a 2D array
y_low = df['Low']
y_high = df['High']
y_close = df['Close']

model_low = LinearRegression()
model_low.fit(X, y_low)

model_high = LinearRegression()
model_high.fit(X, y_high)

model_close = LinearRegression()
model_close.fit(X,y_close)

new_open_values = np.array([[21433.10]])
predicted_low = model_low.predict(new_open_values)
predicted_high = model_high.predict(new_open_values)
predicted_close = model_close.predict(new_open_values)
print('low and high are',predicted_low,predicted_high,predicted_close)
```
